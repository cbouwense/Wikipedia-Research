```python
# Output a string showing the differences between oldtext and newtext.
def showDiff(oldtext, newtext, context=0):
    pwb.PatchManager(oldtext, newtext, context=context).print_hunks()
    
first_str = "apples \n and \n grapes"
second_str = "apples \n but \n frapes"

showDiff(first_str, second_str)
```
