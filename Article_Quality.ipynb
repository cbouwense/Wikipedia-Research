```python
#page = pwb.Page(pwb.Site(), 'High_Sierra_Trail')
#print (getPersistentHistory(page))

def guessQuality(article_vector):
    probablyBad = 0
    if (article_vector['sum'] < 1131):
        probablyBad += 1
    if (article_vector['avg'] < 66):
        probablyBad += 1
    if (article_vector['max'] < 741):
        probablyBad += 1
    if (article_vector['L'] < 5385):
        probablyBad += 1
    if (probablyBad >= 3):
        return 'bad'
    else:
        return 'good'
```

```python
# Get random sample of bad and good quality articles
deleted_sample_size = 15
featured_sample_size = 1

deleted_articles = getRandomArticlesFromCategory('All_articles_proposed_for_deletion', deleted_sample_size)
featured_articles = getRandomArticlesFromCategory('Featured_articles', featured_sample_size)

article_histories = {}

# Timestamp format: %Y-%m-%dT%M-%H-%M-%SZ
starttime = '2017-01-01T00:00:00Z'
endtime = '2018-01-01T00:00:00Z'
#starttime = None
#endtime = None

for article in deleted_articles:
    print ("Getting vector for %s" % article)
    article_histories[article] = getPersistentHistory(article, starttime, endtime)
for article in featured_articles:
    print ("Getting vector for %s" % article)
    article_histories[article] = getPersistentHistory(article, starttime, endtime)
```

```python
quality_guesses = {}
article_count = 1
correct = 0

for article in article_histories:
    quality_guesses[article] = guessQuality(article_histories[article])
    if article_count <= 15 and quality_guesses[article] == 'bad':
        correct += 1
    elif article_count > 15 and quality_guesses[article] == 'good':
        correct += 1
    article_count += 1
    print ("%s: %s" % (article, quality_guesses[article]))
    print ("%s" % article_histories[article])

print ('%s correct out of %s' % (correct, len(article_histories.keys())))
```
